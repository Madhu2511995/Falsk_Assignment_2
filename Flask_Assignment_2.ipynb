### Q1. Explain GET and POST methods.

#### GET method:
The GET method is used to retrieve data from a server. When a client sends a GET request, it includes parameters in the URL itself. 

#### POST method:
The POST method is used to send data to a server to create or update a resource. Unlike the GET method, the data is not included in the URL. Instead, it is sent in the body of the HTTP request. For example, when submitting a form on a website, the form data is sent using a POST request. 

### Q2. Why is request used in Flask?

Request is use to contains data which is send by the client to server.

In Flask, the request object is used to handle incoming HTTP requests from clients. It provides access to the data and information sent by the client, allowing the Flask application to process and respond to the request appropriately

### Q3. Why is redirect() used in Flask?

#### redirect()
By using the redirect() function, Flask provides a straightforward way to manage the navigation and flow of a web application. It enables you to redirect users to different routes or external URLs, handle form submissions, and control access based on authentication or authorization requirements.

### 4. What are templates in Flask? Why is the render_template() function used?

#### Templates

Templates define the structure and layout of web pages by combining HTML markup with placeholders for dynamic content. They allow you to define reusable components, such as headers, footers, navigation menus, and more. Templates provide a way to generate consistent and visually appealing web pages by separating the presentation layer from the underlying application logic.

#### render_template() 
The render_template() function is a built-in Flask function used to render templates and generate HTML responses. It takes the name of the template file as an argument, along with optional additional parameters.
